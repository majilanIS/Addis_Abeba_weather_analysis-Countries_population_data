```python
import pandas as pd
from sklearn.datasets import fetch_openml

df = fetch_openml('titanic', version=1, as_frame=True)['frame']
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
miss_value_per = df.isnull().sum()/len(df)*100
plt.figure(figsize=(10,5))
plt.plot(miss_value_per)
plt.xticks(rotation=90) 
plt.show()
```

```python
plt.figure(figsize=(10,5))
miss_value_per.plot(kind='bar', title='Percentage of missing values', ylabel='Percentage', xlabel='Columns')
```

```python
x, y = df.shape
print(f"Rows: {x}", f"Columns: {y}")
```

```python
df.drop(columns=['body', 'cabin'], inplace=True)
print(f"size of the dataset after dropping columns: {df.shape}")
```

```python
###
# Imputation
###
from sklearn.impute import SimpleImputer
df.isnull()['age'].sum()
```

```python
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print(f"number of missing values in age column after imputation: {df.isnull()['age'].sum()}")
```
