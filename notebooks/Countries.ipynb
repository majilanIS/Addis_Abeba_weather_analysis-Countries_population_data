```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df=pd.read_csv("../data/raw/countries_population_data.csv", sep=",");
    print(" Successfully Loaded Data!")
except:
     print("Failed To Load data!")
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
rows, columns = df.shape
print(f"Rows: {rows}")
print(f"Columns: {columns}")
```

```python
df
```

```python
year_columns = df.columns[4:]  

df[year_columns] = df[year_columns].apply(pd.to_numeric, errors='coerce')

df[year_columns].max().max()
```

```python
df['2024'] = pd.to_numeric(df['2024'], errors='coerce')

df.loc[df['2024'].idxmax(), ['Country Name', '2024']]
```

```python
df_long = df.melt(
    id_vars=['Country Name'],
    value_vars=df.columns[4:],
    var_name='Year',
    value_name='Population'
)
df_long['Population'] = pd.to_numeric(df_long['Population'], errors='coerce')
```

```python
# Extract only needed columns
df_long = df.melt(
    id_vars=['Country Name'],      
    value_vars=df.columns[4:],  
    var_name='Year',
    value_name='Population'
)

# Convert population to numeric
df_long['Population'] = pd.to_numeric(df_long['Population'], errors='coerce')

print(df_long.head())
```

```python
print(df_long['Year'].min())
```

```python
print(df_long['Year'].max())
```

```python
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')

print(df_long['Year'].min())
print(df_long['Year'].max())
```

```python
df_long = df_long[df_long['Year'] >= 2000]
df_long
```

```python
import matplotlib.pyplot as plt

countries = df_long[df_long['Country Name'].isin(['China','India'])]

plt.figure(figsize=(10,5))

for country in countries['Country Name'].unique():
    subset = countries[countries['Country Name'] == country]
    plt.plot(subset['Year'], subset['Population'], label=country)

plt.xlabel("Year")
plt.ylabel("Population")
plt.title("China vs India Population Growth")
plt.legend()
plt.show()
```

```python
# Filter Ethiopia data
ethiopia = df_long[df_long['Country Name'] == 'Ethiopia']


# Plot
plt.figure(figsize=(12,2))

plt.plot(ethiopia['Year'], ethiopia['Population']/10**6)

plt.title("Ethiopia Population from 1960 - 2024")
plt.xlabel("Year")
plt.ylabel("Population")

plt.xticks(rotation=45)
plt.show()
```

```python
countries = df_long[df_long['Country Name'].isin(["Ethiopia", "Kenya"])]

plt.figure(figsize=(10,5))

for country in countries['Country Name'].unique():
    subset = countries[countries['Country Name'] == country]
    plt.plot(subset['Year'], subset['Population']/10**6, label=country)

plt.title("Ethiopia vs Kenya Population")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
ethiopia = df_long[df_long['Country Name'] == 'Ethiopia'].copy()
ethiopia = ethiopia.sort_values('Year')

ethiopia['growth-rate(%)'] = ethiopia['Population'].pct_change()*100
ethiopia[['Year', 'Population', 'growth-rate(%)']].head()
```

```python
plt.figure()
plt.plot(ethiopia['Year'], ethiopia['growth-rate(%)'], color='red')
plt.title("Ethiopia Growth rate (2000-2024)")
plt.xlabel('years')
plt.ylabel('Grawth rate')
plt.show()
```

```python
countries = df_long[df_long['Country Name'].isin(['Ethiopia','Kenya'])]
countries = countries.sort_values(['Country Name','Year'])
countries['Growth Rate (%)'] = countries.groupby('Country Name')['Population'].pct_change() * 100
plt.figure()

for country in countries['Country Name'].unique():
    subset=countries[countries['Country Name'] == country]
    plt.plot(subset['Year'], subset['Growth Rate (%)'], label=country)

plt.title("Ethiopia Vs Kenya Population")
plt.xlabel("years")
plt.ylabel('Growth rate')
plt.legend(['Ethiopia', 'Kenya'])
plt.show()
```
