```python
plt.figure(figsize=(10, 5))
plt.title("Temprature Vs Precipation")
plt.plot(df.temp, df.humidity)
plt.xlabel("Temperature (ÂºC)")
plt.ylabel("Precipitation (mm)")
plt.show()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')

plt.figure(figsize=(5,2))
plt.title("Temprature Over time")
plt.plot(df.temp, df.datetime)
plt.xlabel("Temprature(Â°C)")
plt.ylabel("Date")
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.year 
df['day'] = df['datetime'].dt.day 

df['datetime']
```

```python
plt.figure(figsize=(5,2))
plt.title("Temprature Distribution")
plt.hist(df.temp, bins=20)
plt.xlabel("Temprature (Â°C)")
plt.ylabel("Frequency")
plt.show()
```
