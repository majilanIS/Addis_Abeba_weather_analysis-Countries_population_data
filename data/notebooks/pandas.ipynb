```python
# check what columns are available
print("columns before conversion", df.columns)

# if the datetime column exists, convert it; otherwise, maybe it is already the index
if 'datetime' in df.columns:
    df['datetime'] = pd.to_datetime(df['datetime'])
    df.set_index('datetime', inplace=True)
else:
    print("'datetime' column not found; current index:", df.index)

high_temps = df[df.temp >= df.temp.mean()]
print(high_temps)
```

```python
try:
    __path__="../data/processed/high_temperatures.csv"
    high_temps.to_csv(f"{__path__}", index=True)
    print("Successfully Saved data!")
except:
    print("failed to save data!")
```
